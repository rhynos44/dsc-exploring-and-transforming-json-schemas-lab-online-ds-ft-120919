## Loading the JSON file

Load the data from the file `disease_data.json`.

```python
#Your code here 
import json
import pandas as pd

f = open('disease_data.json')
data = json.load(f)
```

## Convert to a DataFrame

Create a DataFrame from the JSON file. Be sure to retrive the column names for the dataframe. (Search within the 'meta' key of the master dictionary.) The DataFrame should include all 42 columns.

```python
#Your code here
df = pd.DataFrame(data['data'])
cols = [i['name'] for i in data['meta']['view']['columns']]

df.columns = cols
df
```

## Level-Up
## Create a bar graph of states with the highest asthma rates for adults age 18+

```python
df.info()
```

```python
df[df['StratificationCategoryID1'] == 'OVERALL']
```

```python
subset = df[(df['Topic'] == 'Asthma') & (df['StratificationCategoryID1'] == 'OVERALL')]
subset.LocationAbbr.value_counts()
```

```python
subset.Question.value_counts()
```

```python
subset[subset['Question'].str[-11:] == '>= 18 years']
```

```python
subset = df[(df['Topic'] == 'Asthma') & (df['StratificationCategoryID1'] == 'OVERALL') & (df['Question'].str[-11:] == '>= 18 years')]
subset.LocationAbbr.value_counts()
```

```python
subset.DataValueType.value_counts()
```

```python
subset = df[(df['Topic'] == 'Asthma') & (df['StratificationCategoryID1'] == 'OVERALL')
            & (df['Question'].str[-11:] == '>= 18 years') & (df['DataValueType'] == 'Crude Prevalence')]
subset.LocationAbbr.value_counts()
```

```python
cols =['LocationAbbr', 'DataValue']
```

```python
asthma = subset[cols]

asthma.sort_values(by=['DataValue'], ascending=False, inplace=True)
asthma.head()
```

```python
asthma.DataValue = asthma.DataValue.apply(lambda x: float(x))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
asthma.set_index('LocationAbbr')['DataValue'].plot(kind ='bar', figsize=(15,8))
plt.ylabel('Percent Asthma Age >= 18')
```
